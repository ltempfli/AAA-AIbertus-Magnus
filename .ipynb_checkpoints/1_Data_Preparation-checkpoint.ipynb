### Unpack / Format Columns

```python
def format_columns(row):
    
    return pd.Series(data={
        'date' : extract_time(row['date'], row['Time']),
        'temperature' : format_temperature(row['Temperature']),
        'wind_speed' : format_wind(row['Wind Speed']),
        'condition' : row['Condition']
    } , index=None)
    
def extract_time(date_cell, Time_cell):
    
    date = pd.to_datetime(date_cell+"-"+Time_cell, format = '%Y-%m-%d-%I:%M %p')
    
    return date

def format_temperature(temp_cell):
    
    temp_cell = temp_cell.replace('°F', "")
    temp_cell = temp_cell.strip()
    
    return (int(temp_cell)-34) * 5 / 9 # transform Fahrenheit into celsius
    
def format_wind(wind_cell):
    
    return int(wind_cell.replace("mph", "").strip())

def format_precipitation(precip_cell):
    return None
```

```python
weather_data = pd.DataFrame(weather_data_raw.apply(lambda row: format_columns(row), axis=1), index=None);
```
